We know that $\delta \varphi \leq 2 \delta \cdot \max_{x} (\frac{1}{2\sqrt{-(P(0) + x - 1)(P(0) + x)}})$ where $\delta \geq \frac{|\cos(\frac{t}{2} \varphi)|^2 |\sin(\frac{t}{2} \varphi)|^2}{n}$

So, we have a grand total of 3 variables. However we do know something about $\delta$, so let us find the $\max_{x} (\frac{1}{2\sqrt{-(P(0) + x - 1)(P(0) + x)}})$ for different values of $p(0)$

```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def probability_of_estimation_error(delta, n):
    # P [     ]  2e^{2n^2}
    # returns upper bound for probability that the estimation error is greater than delta
    return 2 * np.exp(-2 * n * delta**2)

# Define the color palette
palette = sns.color_palette("bright", len(n))

# plot the probability of estimation error for different values of delta
n = [1, 5, 50, 5000]
delta = np.linspace(-0.00001, 1, 100)

#upper bound
for idx, i in enumerate(n):
    plt.plot(delta, probability_of_estimation_error(delta, i), label=f'n={i}', color=palette[idx])
plt.xlabel(r'$\delta$')
plt.ylabel(r'$\mathbb{P}[|\hat{\theta} - \theta| \geq \delta]$')
plt.xlim(0, 1)
plt.legend()
plt.title(r'Upper bound for $\mathbb{P}[|\hat{\theta} - \theta| \geq \delta]$')
plt.show()

#lower bound
for idx, i in enumerate(n):
    plt.plot(delta, 1 - probability_of_estimation_error(delta, i), label=f'n={i}', color=palette[idx])
plt.xlabel(r'$\delta$')
plt.ylabel(r'$\mathbb{P}[|\hat{\theta} - \theta| < \delta]$')
plt.xlim(0, 1)
plt.legend()
plt.title(r'Lower bound for $\mathbb{P}[|\hat{\theta} - \theta| < \delta]$')
plt.show()
```
